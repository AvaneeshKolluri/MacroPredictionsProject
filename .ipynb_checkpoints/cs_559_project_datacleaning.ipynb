```python
# code added by Thaenraj 
# function to find the percentage of isnull columns; may be required for pre and post nan fix 
# df_merged
def missing_data_info (dataframe) : 
    column_name=[] 
    total_number_missing = [] 
    total_observations  = [] 
    total_missing_percentage  = [] 
    #keep the missing information in a table 
    table_columns = ['dataframe_name', 'column_name', 'total_number_missing', 'total_observations' , 'total_missing_percentage' ] 
    missing_data = pd.DataFrame(columns=table_columns) 
    for col in dataframe.columns:
        column_name.append( col) 
        total_number_missing.append( dataframe[col].isnull().sum())
        total_observations.append(dataframe.shape[0])
        missing_perc = (dataframe[col].isnull().sum()/dataframe.shape[0]) * 100
        total_missing_percentage.append( missing_perc ) 
        #print (dataframe.name , column_name  , total_number_missing , total_observations ,total_missing_percentage )
        #missing_data.loc[len(table_columns )] = [dataframe.name , column_name  , total_number_missing , total_observations ,total_missing_percentage  ]
    missing_data['column_name']  =column_name
    missing_data['total_number_missing']  =total_number_missing
    missing_data['total_observations']  =total_observations
    missing_data['total_missing_percentage']  =total_missing_percentage
    missing_data['dataframe_name']  = 'GDP_data_info'  
    return missing_data 
```

```python
# write a function that takes missing_data dataframe meta data , original dataframe  
# and processes  [ total_number_missing > 0 condition]  matched records 
# use both the interpolation and backwardfill 
# this function should work small data frame that has upto 200 dimensions and 100000 observations 
#def impute_data (missing_data,df_merged):
#print ( missing_data['total_number_missing']  > 0 )
def impute_data (missing_data,df_merged) :
    df_pure = pd.DataFrame() 
    df_pure = missing_data[  missing_data['total_number_missing'] > 0]
    for index, row in df_pure.iterrows() :
        print (row[1])
        df_merged[row[1]]  = df_merged[[row[1]]].interpolate()
        df_merged[row[1]]  = df_merged[[row[1]]].bfill();
    #print(df_merged)
    return df_merged
```
