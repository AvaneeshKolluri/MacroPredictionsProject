```python
from fredapi import Fred
from datetime import datetime, date
from dateutil.relativedelta import relativedelta
import pandas as pd
from pandas_datareader import data as pdr
```

## Get FRED Data

```python
api_key = input("Please input your FRED API Key:\n").strip()
fred = Fred(api_key=api_key)
```

#### GDP

```python
gdp = fred.get_series('GDP')
gdp = pd.DataFrame({"date": gdp.index, "gdp": gdp.values})
gdp
```

```python
unemp = fred.get_series('UNRATE')
unemp = pd.DataFrame({"date": unemp.index, "rate": unemp.values})
unemp
```

```python
sp500 = fred.get_series('SP500')
sp500 = pd.DataFrame({"date": sp500.index, "value": sp500.values})
sp500
```

```python
nasdaq = fred.get_series('NASDAQ100')
nasdaq = pd.DataFrame({"date": nasdaq.index, "value": nasdaq.values})
nasdaq
```

```python
djia = fred.get_series('DJIA')
djia = pd.DataFrame({"date": djia.index, "value": djia.values})
djia
```

```python
quarters = [1, 4, 7, 10]
daily = datetime.today().date() - relativedelta(years=30)
monthly = daily.replace(day=1)
quarterly = monthly.replace(month=min(quarters, key = lambda x : abs(x - daily.month)))
daily, monthly, quarterly
```

```python
gdp = gdp.iloc[gdp.loc[gdp['date'] == str(quarterly)].index[0]:]
gdp
```

```python
unemp = unemp.iloc[unemp.loc[unemp['date'] == str(monthly)].index[0]:]
unemp
```

```python
nasdaq = nasdaq.iloc[nasdaq.loc[nasdaq['date'] == str(daily)].index[0]:]
nasdaq
```

```python
sp500 = pdr.get_data_yahoo('^GSPC', start='1990-01-01', end=date.today())['Adj Close'].to_frame()
sp500
```
